Task: Create an ArrayList of at least 10 integers.

Sort the list in ascending order using Collections.sort().
Filter out all the odd numbers using an enhanced for loop (or with streams if you prefer) and then print the even numbers.
Hint (using streams in Java 8+):

##### Optional:

Display only the odd numbers instead.
Find and print the largest even number from the filtered list.

```python
import java.util.ArrayList;
import java.util.Collections;
import java.util.List;
import java.util.stream.Collectors;

public class PopcornHack {
    public static void main(String[] args) {
        ArrayList<Integer> numbers = new ArrayList<>();
        Collections.addAll(numbers, 23, 12, 45, 8, 4, 19, 33, 16, 2, 28, 7);

        Collections.sort(numbers);

        System.out.println("Even numbers (enhanced for loop):");
        for (int num : numbers) {
            if (num % 2 == 0) {
                System.out.print(num + " ");
            }
        }

        System.out.println("\nOdd numbers (streams):");
        List<Integer> oddNumbers = numbers.stream()
                                          .filter(n -> n % 2 != 0)
                                          .collect(Collectors.toList());
        System.out.println(oddNumbers);

        Integer maxEven = numbers.stream()
                                 .filter(n -> n % 2 == 0)
                                 .max(Integer::compareTo)
                                 .orElse(null);
        System.out.println("Largest even number: " + maxEven);
    }
}
```

Challenge Question (Bonus, +0.01 Points)
Question:
You need to store a collection of student IDs where:

Order does not matter
Duplicate IDs must be prevented
Frequent existence checks are required
Answer:
The most efficient collection to use is a HashSet. This is because a HashSet does not allow duplicate entries, does not maintain order (which is acceptable), and provides constant-time performance for adding, removing, and checking the existence of an element.



```python
import java.util.ArrayList;
import java.util.List;

public class ListExample {
    public static List<Integer> filterEvenNumbers(List<Integer> input) {
        List<Integer> evenList = new ArrayList<>();
        for (int i = 0; i < input.size(); i++) {
            if (input.get(i) % 2 == 0) {
                evenList.add(input.get(i));
            }
        }
        return evenList;
    }

    public static void main(String[] args) {
        List<Integer> nums = new ArrayList<>();
        for (int i = 1; i <= 10; i++) {
            nums.add(i);
        }
        System.out.println("Input List: " + nums);
        System.out.println("Filtered Even Numbers: " + filterEvenNumbers(nums));
    }
}
```

```python
import java.util.HashSet;
import java.util.Set;

public class SetExample {
    public static Set<String> findIntersection(Set<String> set1, Set<String> set2) {
        Set<String> result = new HashSet<>();
        for (String item : set1) {
            if (set2.contains(item)) {
                result.add(item);
            }
        }
        return result;
    }

    public static void main(String[] args) {
        Set<String> set1 = new HashSet<>();
        Set<String> set2 = new HashSet<>();

        set1.add("apple");
        set1.add("banana");
        set1.add("cherry");
        set1.add("date");

        set2.add("banana");
        set2.add("date");
        set2.add("fig");
        set2.add("grape");

        System.out.println("Set1: " + set1);
        System.out.println("Set2: " + set2);
        System.out.println("Intersection: " + findIntersection(set1, set2));
    }
}
```

```python
import java.util.ArrayDeque;
import java.util.Deque;

public class DequeExample {
    public static void main(String[] args) {
        Deque<String> line = new ArrayDeque<>();

        // Enqueue: add 4 customers to the end
        line.addLast("Alice");
        line.addLast("Bob");
        line.addLast("Charlie");
        line.addLast("Diana");
        System.out.println("Initial line after adding 4 customers (end): " + line);

        // VIP Arrival
        line.addFirst("VIP1");
        System.out.println("After VIP arrival (added at front): " + line);

        // Process Line: remove front
        line.removeFirst();
        System.out.println("After serving the customer at the front (removed): " + line);

        // Inspect Line
        System.out.println("Current front of the line: " + line.peekFirst());
        System.out.println("Current back of the line: " + line.peekLast());
        System.out.println("Total number of customers waiting: " + line.size());
    }
}
```
