#### Unit 3.5 Compound booleans
Popcorn Hack 1

explain the purpose of this algorithm, and what each if condition is used for
what would be output if input is
- age 20
- anual income 1500
- student status: yes

```python
function checkMembershipEligibility() {
    // Get user input
    let age = parseInt(prompt("Enter your age:"));  // Age input
    let income = parseFloat(prompt("Enter your annual income:"));  // Annual income input
    let isStudent = prompt("Are you a student? (yes/no):").toLowerCase() === 'yes';  // Student status input

    // Initialize an empty array to hold results
    let results = [];

    // Check eligibility for different memberships

    // Basic Membership
    if (age >= 18 && income >= 20000) {
        results.push("You qualify for Basic Membership.");
    }

    // Premium Membership
    if (age >= 25 && income >= 50000) {
        results.push("You qualify for Premium Membership.");
    }

    // Student Discount
    if (isStudent) {
        results.push("You are eligible for a Student Discount.");
    }

    // Senior Discount
    if (age >= 65) {
        results.push("You qualify for a Senior Discount.");
    }

    // If no eligibility, provide a default message
    if (results.length === 0) {
        results.push("You do not qualify for any memberships or discounts.");
    }

    // Output all results
    results.forEach(result => console.log(result));
}

// Call the function to execute
checkMembershipEligibility();
```

- The algorithm checks for Membership eligibility and the if statements check for membership and discount based on age and if you are a student.
- For the given inputs the program will return "You are eligible for a Student Discount"

#### Homework: Membership Recommendation System


```python
import java.util.Scanner;

public class PrimeClubRecommendation {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        int age = 0;
        double annualIncome = 0.0;
        String studentStatus;
        String employmentType;

        // Input validation for age
        while (true) {
            System.out.print("Enter your age (positive integer): ");
            if (scanner.hasNextInt()) {
                age = scanner.nextInt();
                if (age > 0) break;
                else System.out.println("Age must be a positive integer.");
            } else {
                System.out.println("Invalid input. Please enter a valid positive integer for age.");
                scanner.next(); // Clear invalid input
            }
        }

        // Input validation for annual income
        while (true) {
            System.out.print("Enter your annual income (positive number): ");
            if (scanner.hasNextDouble()) {
                annualIncome = scanner.nextDouble();
                if (annualIncome > 0) break;
                else System.out.println("Annual income must be a positive number.");
            } else {
                System.out.println("Invalid input. Please enter a valid positive number for annual income.");
                scanner.next(); // Clear invalid input
            }
        }

        // Input validation for student status
        while (true) {
            System.out.print("Are you a student? (yes/no): ");
            studentStatus = scanner.next().trim().toLowerCase();
            if (studentStatus.equals("yes") || studentStatus.equals("no")) {
                break;
            } else {
                System.out.println("Invalid input. Please enter 'yes' or 'no' for student status.");
            }
        }

        // Input validation for employment type
        while (true) {
            System.out.print("Enter your employment type (full-time, part-time, unemployed): ");
            employmentType = scanner.next().trim().toLowerCase();
            if (employmentType.equals("full-time") || employmentType.equals("part-time") || employmentType.equals("unemployed")) {
                break;
            } else {
                System.out.println("Invalid input. Please enter 'full-time', 'part-time', or 'unemployed' for employment type.");
            }
        }

        // Membership qualification logic
        String recommendation = null;

        if (age >= 25 && annualIncome >= 50000) {
            recommendation = "Premium Membership";
        } else if (age >= 65) {
            recommendation = "Senior Discount";
        } else if (studentStatus.equals("yes")) {
            recommendation = "Student Discount";
        } else if (age >= 18 && annualIncome >= 20000) {
            recommendation = "Basic Membership";
        }

        // Print final recommendation
        if (recommendation != null) {
            System.out.println("The best membership you qualify for is: " + recommendation);
        } else {
            System.out.println("You do not qualify for any memberships or discounts.");
        }

        scanner.close();
    }
}
```
